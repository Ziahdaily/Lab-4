# Dataset Preperation 1

One of the two datasets that I chose to use is attendance. For this dataset I have decided to rename all of the columns.

```python
# this code deletes the first four rows fo code by using drop.

attendance_df = attendance_df.drop(attendance_df.index[0:3])

# afterwards I want to see if the code, worked so I print the data
attendance_df.head(6)
```

```python
# new names for the columns/headers

attendance_df.rename(columns ={
    attendance_df.columns[0]:"Place",
    attendance_df.columns[1]:"ADA as percent of enrollment", 
    attendance_df.columns[2]:"NaN",
    attendance_df.columns[3]:"Average days in school year", 
    attendance_df.columns[4]:"NaN",
    attendance_df.columns[5]:"ADA as percent of enrollment-Elementary schools", 
    attendance_df.columns[6]:'NaN',
    attendance_df.columns[7]:"ADA as percent of enrollment-Secondary schools",
    attendance_df.columns[8]:"NaN",
    attendance_df.columns[9]:"Average hours in school day-Secondary schools",
    attendance_df.columns[10]:"NaN",
    attendance_df.columns[11]:"Average hours in school day-Secondary schools",
    attendance_df.columns[12]:"NaN",
    attendance_df.columns[13]:"Average hours in school day-Secondary schools",
    attendance_df.columns[14]:"NaN",
    attendance_df.columns[15]:"Average hours in school day-Secondary schools",
    attendance_df.columns[16]:"NaN"
    },

            
        inplace = True)

#Now I want to print the first 9 rows to make sure that the code worked.
attendance_df.head(9)
```

```python
#For the code below I want to drop some all of the collumns witht he name NaN. I do this becuase if I dont know what the data is about it isnt relevent and therefore should be dropped.
attendance_df = attendance_df.drop("NaN",axis=1)

#Then I want to print the first 9 ros to make sure that my code worked
attendance_df.head(9)
```

```python
# new names for the columns/headers

attendance_df.rename(columns ={
    attendance_df.columns[0]:"Place",
    attendance_df.columns[1]:"ADA as percent of enrollment", 
    attendance_df.columns[2]:"Average hours in school day",
    attendance_df.columns[3]:"Average days in school year", 
    attendance_df.columns[4]:"Average hours in school year",
    attendance_df.columns[5]:"ADA as percent of enrollment-Elementary schools", 
    attendance_df.columns[6]:'Average hours in school day-Elementary schools',
    attendance_df.columns[7]:"ADA as percent of enrollment-Secondary schools",
    attendance_df.columns[7]:"Average hours in school day-Secondary schools"
    },
            
        inplace = True)

#Now I want to print the first 9 rows of data to mak sure that thee code above works
attendance_df.head(9)
```
